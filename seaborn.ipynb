## An introduction to seaborn
* Seaborn is a library for making statistical graphics in Python.
* It builds on top of matplotlib and integrates closely with pandas data structures.
* Seaborn helps you explore and understand your data.

```python
import pandas as pd
import matplotlib. pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset('titanic')
```

## Titanic Dataset Features:
* survived:Whether the passenger survived or not (binary: 0 = No, 1 = Yes).
* pclass:Passenger class (integer: 1 = 1st, 2 = 2nd, 3 = 3rd).
* sex:Gender of the passenger (male/female).
* age:Age of the passenger (float). Some values are missing and appear as NaN.
* sibsp:Number of siblings or spouses aboard the Titanic (integer).
* parch:Number of parents or children aboard the Titanic (integer).
* fare:Ticket fare paid by the passenger (float).
* embarked:Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).
* class:A categorical version of the pclass feature (1st, 2nd, 3rd class).
* who:Describes whether the passenger is male, female, or a child (man/woman/child).
* adult_male:Whether the passenger is an adult male (boolean: True/False).
* deck:Deck level where the passenger’s cabin is located (A, B, C, D, E, F, G, T, or NaN if unknown).
* embark_town:The town where the passenger embarked (Cherbourg, Queenstown, Southampton).
* alive:Whether the passenger survived or not (yes/no).
* alone:Whether the passenger was alone (boolean: True if alone, False if with family).

## COUNT PLOT
* Displays the count of observations in each categorical bin.
* Similar to a bar plot, but specifically counts the occurrences of each category. Useful for categorical data analysis.

```python
sns.countplot(x=data['sex'])
plt.show()
```

```python
sns.countplot(x=data['class'])
plt.title('PASSENGERS CLASS DISTRIBUTION',color='violet')
plt.show()
```

```python
sns.countplot(x=data['embark_town'],color='green')
plt.title('EMBARKED TOWN OF PASSENGERS')
plt.show()
```

## BAR PLOT

```python
sns.barplot(x='who',y='fare',data=data)
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=data,hue='class')
plt.show()
```

```python
sns.barplot(x='pclass',y='survived',data=data,hue='sex');
```

## PIE PLOT

```python
data['survived'].value_counts().plot(kind='pie',title='survived deatils of passengers',explode=(0,0.1),autopct="%1.2f%%");
```

```python
data['class'].value_counts().plot(kind='pie',title='PASSENGERS CLASS DISTRIBUTION',explode=(0,0,0.1),shadow=True)
plt.show()
```

```python
sns.scatterplot(x='class',y='fare',hue='sex',data=data)
plt.show()
```
